This notebook utilizes data taken by an ESP8266 microcontroller via an MS5803-05BA sensor during freezing experiments to explore how pressure and temperature change throughout the freezing process.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16, 'axes.labelsize': 18, 'legend.fontsize': 14})
```

```python
# load all data
# 10-19-23
delimiters = r'[:,]'
water_data_10_19 = pd.read_csv('Pressure Data/-30_10-19-23/water_data.csv',sep=delimiters,header=None,engine='python')
length_10_19 = water_data_10_19.index.tolist()

# 10-23-23
delimiters = r'[:,]'
water_data_10_23 = pd.read_csv('Pressure Data/-30_10-23-23/water_data.csv',sep=delimiters,header=None,engine='python')
length_10_23 = water_data_10_23.index.tolist()

# 10-25-23
delimiters = r'[:,]'
water_data_10_25 = pd.read_csv('Pressure Data/-40_10-25-23/water_data.csv',sep=delimiters,header=None,engine='python')
length_10_25 = water_data_10_25.index.tolist()

# 10-26-23 Not plotting because ~40 min of fracture data was not taken due to errors
delimiters = r'[:,]'
water_data_10_26 = pd.read_csv('Pressure Data/-40_10-26-23_major_errors/water_data.csv',sep=delimiters,header=None,engine='python')
length_10_26 = water_data_10_26.index.tolist()

# 11-6-23
delimiters = r'[:,]'
water_data_11_6 = pd.read_csv('Pressure Data/-30_11-6-23/water_data.csv',sep=delimiters,header=None,engine='python')
length_11_6 = water_data_11_6.index.tolist()

# 11-15-23 CONTROL (SENSOR IS IN AIR)
delimiters = r'[:,]'
control = pd.read_csv('Pressure Data/-30_11-15-23 control/water_data.csv',sep=delimiters,header=None,engine='python')
control_length = control.index.tolist()

# 11-28-23 SALT 6 g/l
delimiters = r'[:,]'
salt_6_11_28 = pd.read_csv('Pressure Data/-30_11-28-23_salt_6_l/water_data.csv',sep=delimiters,header=None,engine='python')
salt_6_11_28_length = salt_6_11_28.index.tolist()
```

## Calculate running mean of pressure

```python
plt.rcParams.update({'font.size': 20, 'axes.labelsize': 26, 'legend.fontsize': 20})
# List of window sizes
window_sizes = [1, 5, 10, 20,30,40]

# Create a grid of subplots
fig, axes = plt.subplots(len(window_sizes), 2, sharey=True, figsize=(22,30))

# Loop through each window size and plot the results
for i, window_size in enumerate(window_sizes):
    salt_running = salt_6_11_28[1].rolling(window=window_size).mean()
    water_running = water_data_11_6[1].rolling(window=window_size).mean()

    # Plot 1
    axes[i, 0].plot((np.array(salt_6_11_28_length) * 30) / 60 / 60, salt_running, 'b', label='pressure')
    axes[i, 0].set_xlabel('hours', fontsize=12)
    axes[i, 0].set_ylabel('millibars', color='b', fontsize=12)
    axes[i, 0].tick_params('y', colors='b', labelsize=8)
    axes[i, 0].set_xticks(np.arange(0, len(length_11_6), step=1))
    axes[i, 0].set_xlim(0, 5)
    axes[i, 0].set_ylim(0, 3500)
    axes[i, 0].set_title(f'Running Mean Salty (Window Size {window_size})')

    ax2_1 = axes[i, 0].twinx()
    ax2_1.plot((np.array(salt_6_11_28_length) * 30) / 60 / 60, salt_6_11_28[3], 'r', label='Temperature', linestyle='--')
    ax2_1.set_ylabel('°C', color='r', fontsize=12)
    ax2_1.tick_params('y', colors='r', labelsize=8)

    axes[i, 0].legend(loc='lower left', fontsize=8, bbox_to_anchor=(0, 0.08))
    ax2_1.legend(loc='lower left', fontsize=8)

    # Plot 2
    axes[i, 1].plot((np.array(length_11_6) * 30) / 60 / 60, water_running, 'b', label='pressure')
    axes[i, 1].set_xlabel('hours', fontsize=12)
    axes[i, 1].set_xticks(np.arange(0, len(length_11_6), step=1))

    ax2_2 = axes[i, 1].twinx()
    ax2_2.plot((np.array(length_11_6) * 30) / 60 / 60, water_data_11_6[3], 'r', label='Temperature', linestyle='--')
    ax2_2.set_ylabel('°C', color='r', fontsize=12)
    ax2_2.tick_params('y', colors='r', labelsize=8)

    axes[i, 1].legend(loc='lower left', fontsize=8, bbox_to_anchor=(0, 0.08))
    ax2_2.legend(loc='lower left', fontsize=8)

    axes[i, 1].set_xlim(0, 5)
    axes[i, 1].set_title(f'Running Mean Fresh (Window Size {window_size})')

plt.tight_layout()
plt.show()
```

## Calculate time derivative of pressure

```python
salt_6_11_28_dp_dt = np.gradient(salt_6_11_28[1], (np.array(salt_6_11_28_length) * 30))
water_data_11_6_dp_dt = np.gradient(water_data_11_6[1], (np.array(length_11_6) * 30))
```

## try fourier transforms of dp/dt and p

```python
# Sample data for water dp/dt
time_water_dp_dt = np.arange(0, len(water_data_11_6_dp_dt)) * 30  # Assuming data is sampled every 30 seconds
pressure_water_dp_dt = water_data_11_6_dp_dt

# FFT calculation for water dp/dt
fft_result_water_dp_dt = np.fft.fft(pressure_water_dp_dt)
freq_water_dp_dt = np.fft.fftfreq(len(pressure_water_dp_dt), d=30)  # d is the sample spacing (30 seconds)

# Sample data for water pressure
time_water = np.arange(0, len(water_data_11_6)) * 30  # Assuming data is sampled every 30 seconds
pressure_water = water_data_11_6[1]

# FFT calculation for water pressure
fft_result_water = np.fft.fft(pressure_water)
freq_water = np.fft.fftfreq(len(pressure_water), d=30)  # d is the sample spacing (30 seconds)

# Plotting
plt.figure(figsize=(16, 8))

# Subplot 1: Original signal for water dp/dt
plt.subplot(2, 2, 1)
plt.plot(time_water_dp_dt, pressure_water_dp_dt, label='Water dp/dt - Original Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('dp/dt')
plt.legend()

# Subplot 2: FFT result for water dp/dt
plt.subplot(2, 2, 2)
plt.plot(freq_water_dp_dt, np.abs(fft_result_water_dp_dt))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Water dp/dt - FFT Result')

# Subplot 3: Original signal for water pressure
plt.subplot(2, 2, 3)
plt.plot(time_water, pressure_water, label='Water - Original Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Pressure')
plt.legend()

# Subplot 4: FFT result for water pressure
plt.subplot(2, 2, 4)
plt.plot(freq_water, np.abs(fft_result_water))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Water - FFT Result')

plt.tight_layout()
plt.show()
```
