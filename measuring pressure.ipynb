This notebook utilizes data taken by an ESP8266 microcontroller via an MS5803-05BA sensor during freezing experiments to explore how pressure and temperature change throughout the freezing process.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load all data
# 10-19-23
delimiters = r'[:,]'
water_data_10_19 = pd.read_csv('Pressure Data/-30_10-19-23/water_data.csv',sep=delimiters,header=None,engine='python')
length_10_19 = water_data_10_19.index.tolist()

# 10-23-23
delimiters = r'[:,]'
water_data_10_23 = pd.read_csv('Pressure Data/-30_10-23-23/water_data.csv',sep=delimiters,header=None,engine='python')
length_10_23 = water_data_10_23.index.tolist()

# 10-25-23
delimiters = r'[:,]'
water_data_10_25 = pd.read_csv('Pressure Data/-40_10-25-23/water_data.csv',sep=delimiters,header=None,engine='python')
length_10_25 = water_data_10_25.index.tolist()

# 10-26-23 Not plotting because ~40 min of fracture data was not taken due to errors
delimiters = r'[:,]'
water_data_10_26 = pd.read_csv('Pressure Data/-40_10-26-23_major_errors/water_data.csv',sep=delimiters,header=None,engine='python')
length_10_26 = water_data_10_26.index.tolist()

# 11-6-23
delimiters = r'[:,]'
water_data_11_6 = pd.read_csv('Pressure Data/-30_11-6-23/water_data.csv',sep=delimiters,header=None,engine='python')
length_11_6 = water_data_11_6.index.tolist()

# 11-15-23 CONTROL (SENSOR IS IN AIR)
delimiters = r'[:,]'
control = pd.read_csv('Pressure Data/-30_11-15-23 control/water_data.csv',sep=delimiters,header=None,engine='python')
control_length = control.index.tolist()

# 11-28-23 SALT 6 g/l
delimiters = r'[:,]'
salt_6_11_28 = pd.read_csv('Pressure Data/-30_11-28-23_salt_6_l/water_data.csv',sep=delimiters,header=None,engine='python')
salt_6_11_28_length = salt_6_11_28.index.tolist()
```

```python
plt.rcParams.update({'font.size': 16})
fig, ax1 = plt.subplots()
ax1.plot((np.array(salt_6_11_28_length) * 30) / 60 / 60, salt_6_11_28[1], 'b', label='Pressure')
ax1.set_xlabel('hours',fontsize=22)
ax1.set_ylabel('millibars', color='b', fontsize = 22)
ax1.tick_params('y', colors='b', labelsize=12)
ax1.set_xticks(np.arange(0, len(length_11_6), step=1))
plt.ylim(0,3500)

ax2 = ax1.twinx()
ax2.plot((np.array(salt_6_11_28_length) * 30) / 60 / 60, salt_6_11_28[3], 'r', label='Temperature', linestyle = '--')
ax2.set_ylabel('°C', color='r', fontsize =22)
ax2.tick_params('y', colors='r', labelsize=12)

ax1.legend(loc='lower left', fontsize=12, bbox_to_anchor=(0, 0.08))
ax2.legend(loc='lower left', fontsize=12)
#plt.suptitle('11/6/23', fontsize = 14)
#plt.title('Salted 6g/l Frozen at -30 °C',fontsize=22)
plt.xlim(0,5)
#print('note that this sample was not water or air tight as evidenced by drips through the pressure sensor seal')
#plt.savefig('Pressure Temperature FirstTrial.png', dpi=300, bbox_inches='tight')
plt.savefig('Pressure_Salt_6g 11-28-23.png', dpi=450, bbox_inches='tight')
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})
fig, ax1 = plt.subplots()
ax1.plot((np.array(length_11_6) * 30) / 60 / 60, water_data_11_6[1], 'b', label='Pressure')
ax1.set_xlabel('hours',fontsize=22)
ax1.set_ylabel('millibars', color='b', fontsize = 22)
ax1.tick_params('y', colors='b', labelsize=12)
ax1.set_xticks(np.arange(0, len(length_11_6), step=1))

ax2 = ax1.twinx()
ax2.plot((np.array(length_11_6) * 30) / 60 / 60, water_data_11_6[3], 'r', label='Temperature', linestyle = '--')
ax2.set_ylabel('°C', color='r', fontsize =22)
ax2.tick_params('y', colors='r', labelsize=12)

ax1.legend(loc='lower left', fontsize=12, bbox_to_anchor=(0, 0.08))
ax2.legend(loc='lower left', fontsize=12)
#plt.suptitle('11/6/23', fontsize = 14)
#plt.title('Unsalted Frozen at -30 °C',fontsize=22)
plt.xlim(0,4)
#print('note that this sample was not water or air tight as evidenced by drips through the pressure sensor seal')
#plt.savefig('Pressure Temperature FirstTrial.png', dpi=300, bbox_inches='tight')
plt.savefig('Pressure Temperature 11-6-23.png', dpi=450, bbox_inches='tight')
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})
fig, ax1 = plt.subplots()
ax1.plot(length_10_19, water_data_10_19[1], 'b', label='Pressure')
ax1.set_xlabel('Measurement # (~1 Measurement/30 sec)',fontsize=14)
ax1.set_ylabel('millibars', color='b')
ax1.tick_params('y', colors='b', labelsize=12)

ax2 = ax1.twinx()
ax2.plot(length_10_19, water_data_10_19[3], 'r', label='Temperature')
ax2.set_ylabel('°C', color='r')
ax2.tick_params('y', colors='r', labelsize=12)

ax1.legend(loc='lower left', fontsize=12)
ax2.legend(loc='upper right', fontsize=12)
plt.title('10-19-23 Pressure and Temperature by Measurement #')
plt.suptitle('Frozen at -30 °C',fontsize=12)
plt.xlim()
#plt.savefig('Pressure Temperature FirstTrial.png', dpi=300, bbox_inches='tight')
plt.show()
```
