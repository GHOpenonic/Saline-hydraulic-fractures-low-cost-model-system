```python
avg_curve_x = []
for i in range(0, len(curve_fractured_x)):
    difference = (curve_not_fractured_x[i + 1] - curve_fractured_x[i])
    mid_length= difference / 2
    midpoint = mid_length + curve_fractured_x[i]
    avg_curve_x.append(midpoint)
avg_curve_x
```
